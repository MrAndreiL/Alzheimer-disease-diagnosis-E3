```python
def get_subject_id(path):
    return re.search(r'OAS(\d{5})',path).group(0)

def get_day(path):
    return int(re.search(r'd(\d{4})',path).group(0)[1:])

def get_subject_days_and_diagnosis(subject_id):
    days = []
    for x in subjects.loc[subjects['Subject'] == subject_id]["Label"].tolist():
        if match := pattern.search(x):
            days.append(int(match.group(0)))
    diagonistics = subjects.loc[subjects['Subject'] == subject_id]["dx1"].tolist()
    return days,diagonistics

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def get_diagnostic_base_on_day(path):
    a,b = get_subject_days_and_diagnosis(get_subject_id(path))
    get_day(path)
    return b[a.index(find_nearest(a,757))]
```

```python
# for p in paths:
#     print(get_diagnostic_base_on_day(p))
def get_valid_filename(s):
    s = str(s).strip().replace(' ', '_')
    return re.sub(r'(?u)[^-\w.]', '', s)
```
